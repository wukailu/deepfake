```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
plt.style.use('ggplot')
import sys
import os
sys.path.append("..")
sys.path.append(".")
sys.path.append("./utils")
print(sys.path)
import insightface
import cv2
import matplotlib.pylab as plt
import face_recognition
from tqdm import tqdm
from PIL import Image, ImageDraw
from skvideo.io import vread
from insightface.app.face_analysis import  Face
import random
from settings import DATA_DIR, face_cache_path
from engine import Cache_loader
```

```python
meta = pd.read_json(os.path.join("/data/deepfake/dfdc_train",'metadata_all.json')).T
```

```python
def dist(emb1, emb2):
    return np.sum(np.square(emb1 - emb2))

def get_embedding(index):
    A = insightface.app.FaceAnalysis()
    A.prepare(0)
    
    id2feature = dict()
    for fn in tqdm(index):
        path = os.path.join(face_cache_path, fn)
        frames = Cache_loader.extract_video(path)
        for fid, image in frames.items():
            face_list = A.get(image)
            if len(face_list) == 0:
                print(f'Could not find face in frame {fid} of {fn}')
                continue
            id2feature[(fn, fid)] = face_list[0].normed_embedding
    return id2feature

ret = get_embedding(meta.index[:10])
ret
```

```python
id2feature = get_embedding(meta.index)
```

```python
feature_df = pd.DataFrame([[key[0],key[1],it] for key,it in id2feature.items()], columns=['filename','frame','feature'])
```

```python
feature_df["label"] = feature_df["filename"].apply(lambda x: meta.at[x, "label"])
feature_df["original"] = feature_df["filename"].apply(lambda x: meta.at[x, "original"])
```

# 为每个fake寻找最近的real

```python
real_df = feature_df[feature_df["label"]=="REAL"]
real_features = np.array(list(real_df["feature"]))
real_fn, frames = list(real_df["filename"]), list(real_df["frame"])

fake_df = feature_df[feature_df["label"]=="FAKE"]

fake_set = {}
for index, row in fake_df.iterrows():
    fake_set[row["filename"]]=row["feature"]
fake_features = np.array([fea for fn, fea in fake_set.items()])
fake_fn = [fn for fn, fea in fake_set.items()]
# fake_features = np.array(list(fake_df["feature"]))
# fake_fn = list(fake_df["filename"])
```

```python
from sklearn.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(real_features)
```

```python
distances, indices = nbrs.kneighbors(fake_features)
fake2real = {fake_fn[idx]:real_fn[i[0]] for idx, i in enumerate(indices)}
```

```python
import pickle
save_path = "/home/kailu/deepfake/preprocess/fake2real.pkl"
with open(save_path, 'wb') as handle:
    pickle.dump(fake2real, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
# with open(save_path, 'rb') as file:
#     fake2real = pickle.load(file)
```

```python
metadata = pd.read_json("/data/deepfake/metadata/metadata_40_49_newdrop.json").T
```

```python
def clean(row):
    if row.label == "REAL":
        return True
    if row.name in fake2real.keys():
        fr1 = metadata.at[fake2real[row.name], "split_kailu"]=="train" 
        fr2 = metadata.at[row.original, "split_kailu"]
        if (fr1 == "train" and fr2 == "validation") or (fr1 == "train" and fr2 == "test"):
            return False
        return True
    else:
        return False
    
select = metadata.apply(clean,1)
```
